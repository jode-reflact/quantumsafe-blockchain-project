```python
import os
import pandas as pd
import numpy as np
import json
```

```python
aggregated_results_path = os.path.join(os.getcwd(), "results", "aggregated.csv")
aggregated_results_df = pd.read_csv(aggregated_results_path)
```

```python
def calcMedianTime(dist_block_time):
    temp = json.loads(dist_block_time)
    all = list()
    for block_time in temp:
        for time in block_time:
            all.append(time)
    return np.median(all)
```

```python
aggregated_results_df["median_block_time"] = aggregated_results_df["dist_block_time"].apply(calcMedianTime)
```
