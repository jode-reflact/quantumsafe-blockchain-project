```python
import requests
```

```python
def get_blocks():
  endpoint = "http://localhost:2001/blocks"

  res = requests.get(endpoint)
  return res.json()
```

```python
def add_block():
  endpoint = "http://localhost:2001/blocks"

  block = {
    "index": 1
  }

  res = requests.post(endpoint,
                      block)
  return res
```
